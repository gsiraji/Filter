```python
import scipy as sp
import numpy as np
import pandas as pd
```

```python
eeg = pd.read_csv("data/sample_EEG_data.csv")

bg = pd.read_csv("data/60Hz_notch_FIR_weights.csv")


eeg = np.asarray(eeg)

bg = np.asarray(bg)
```

```python
c1 = sp.signal.convolve(eeg,bg,mode="full",method="direct")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(16,2,figsize=(10,15))
for i in range(16):
    ax[i,0].plot(eeg[0:30000:50,i:i+1],':.')
    ax[i,1].plot(c1[0:30000:50,i:i+1],':.')
```
